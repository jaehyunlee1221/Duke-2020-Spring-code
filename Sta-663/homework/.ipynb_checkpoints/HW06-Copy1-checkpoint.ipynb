```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
```

```python
def logistic(x):
    """Logistic function."""
    return np.exp(x)/(1 + np.exp(x))

def gd(X, y, beta, alpha, niter):
    """Gradient descent algorihtm."""
    n, p = X.shape
    Xt = X.T
    for i in range(niter):
        y_pred = logistic(X @ beta)
        epsilon = y - y_pred
        grad = Xt @ epsilon / n
        beta += alpha * grad
    return beta
```

```python
x = np.linspace(-6, 6, 100)
plt.plot(x, logistic(x))
pass
```

```python
n = 10000
p = 2
X, y = make_blobs(n_samples=n, n_features=p, centers=2, 
                  cluster_std=1.05, random_state=23)
X = np.c_[np.ones(len(X)), X]
y = y.astype('float')
```

```python
# initial parameters
niter = 1000
α = 0.01
β = np.zeros(p+1)

# call gradient descent
β = gd(X, y, β, α, niter)

# assign labels to points based on prediction
y_pred = logistic(X @ β)
labels = y_pred > 0.5

# calculate separating plane
sep = (-β[0] - β[1] * X)/β[2]

plt.scatter(X[:, 1], X[:, 2], c=labels, cmap='winter')
plt.plot(X, sep, 'r-')
pass
```

```python
import numba
from numba import jit
from numba import int32, int64, float32, float64
```

```python
@numba.vectorize([float64(float64)])

def logistic_numba_cpu(x):
    """Logistic function."""
    return np.exp(x)/(1 + np.exp(x))
```

```python
@numba.vectorize([float64(float64)],target = "parallel")

def logistic_numba_parallel(x):
    """Logistic function."""
    return np.exp(x)/(1 + np.exp(x))
```

```python
np.testing.assert_array_almost_equal(logistic(x),logistic_numba_cpu(x))
np.testing.assert_array_almost_equal(logistic(x),logistic_numba_parallel(x))
```

```python
@jit(float64[:](float64[:]))
def logistic_numba(x):
    """Logistic function."""
    return np.exp(x)/(1 + np.exp(x))
```

```python
@jit(float64[:](float64[:,::1], float64[:],float64[::1],float64, int32))
def gd_numba(X, y, beta, alpha, niter):
    """Gradient descent algorihtm."""
    n, p = X.shape
    Xt = X.T
    for i in range(niter):
        y_pred = logistic_numba(X @ beta)
        epsilon = y - y_pred
        grad = Xt @ epsilon / n
        beta += alpha * grad
    return beta
```

```python
# initial parameters
niter = 1000
α = 0.01
β = np.zeros(p+1)

# call gradient descent
β = gd(X, y, β, α, niter)
b = gd_numba(X, y, β, α, niter)
```

```python
%%cython -a

import cython
import numpy as np
cimport numpy as np
from math import exp

@cython.boundscheck(False)
@cython.wraparound(False)
def logistic_cython(double[:] x):
    """Logistic function."""
    cdef int i
    cdef int n = x.shape[0]
    cdef double [:] res = np.empty(n)
    
    for i in range(n):
        res[i] = exp(x[i])/(1.0 + exp(x[i]))
    return res
```

```python
np.testing.assert_array_almost_equal(logistic(x),logistic_cython(x))
```

```python
%timeit -r3 -n5 logistic(x)
%timeit -r3 -n5 logistic_numba_cpu(x)
%timeit -r3 -n5 logistic_numba(x)
%timeit -r3 -n5 logistic_cython(x)
```

```python
%%cython -a 

import cython
import numpy as np
cimport numpy as np
from math import exp

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
cdef double[:] logistic_cython(double[:] x):
    """Logistic function."""
    cdef int i
    cdef int n = x.shape[0]
    cdef double [:] res = np.empty(n)
    
    for i in range(n):
        res[i] = exp(x[i])/(1.0 + exp(x[i]))
    return res

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
def gd_cython(double [:,:] X, double[:] y, double[:] beta, double alpha, int niter):
    """Gradient descent algorihtm."""
    cdef int n = X.shape[0]
    cdef int p = X.shape[1]
    cdef int i,j,k
    cdef double[:,:] Xt = X.T
    cdef double[:] y_pred = np.empty(n)
    cdef double[:] epsilon = np.empty(n)
    cdef double[:] grad = np.empty(p)
    
    for i in range(niter):
        y_pred = logistic_cython(np.dot(X,beta))
        for j in range(n):
            epsilon[j] = (y[j] - y_pred[j])/n
        grad = np.dot(Xt,epsilon)
        for k in range(p):
            beta[k] += alpha*grad[k]
    return beta
```

```python
np.testing.assert_array_almost_equal(gd(X, y, β, α, niter),gd_cython(X, y, β, α, niter))
```

```python
%timeit -r3 -n5 gd(X, y, β, α, niter)
%timeit -r3 -n5 gd_numba(X, y, β, α, niter)
%timeit -r3 -n5 gd_cython(X, y, β, α, niter)
```

```python
%%cython -a 

import cython
import numpy as np
cimport numpy as np
from math import exp

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
cdef double[:] logistic_cython(double[:] x):
    """Logistic function."""
    cdef int i
    cdef int n = x.shape[0]
    cdef double [:] res = np.empty(n)
    
    for i in range(n):
        res[i] = exp(x[i])/(1.0 + exp(x[i]))
    return res

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
cdef double[:] dot(double[:,:] X, double[:] v):
    cdef int n = X.shape[0]
    cdef int p = X.shape[1]
    cdef int i,j
    cdef double d
    cdef double[:] res = np.empty(n)
    
    for i in range(n):
        d = 0.0
        for j in range(p):
            d += X[i,j]*v[j]
        res[i] = d
    return res

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
def gd_cython2(double [:,:] X, double[:] y, double[:] beta, double alpha, int niter):
    """Gradient descent algorihtm."""
    cdef int n = X.shape[0]
    cdef int p = X.shape[1]
    cdef int i,j,k
    cdef double[:,:] Xt = X.T
    cdef double[:] y_pred = np.empty(n)
    cdef double[:] epsilon = np.empty(n)
    cdef double[:] grad = np.empty(p)
    
    for i in range(niter):
        y_pred = logistic_cython(dot(X,beta))
        for j in range(n):
            epsilon[j] = (y[j] - y_pred[j])/n
        grad = dot(Xt,epsilon)
        for k in range(p):
            beta[k] += alpha*grad[k]
    return beta
```

```python
np.testing.assert_array_almost_equal(gd(X, y, β, α, niter),gd_cython2(X, y, β, α, niter))
%timeit -r3 -n5 gd(X, y, β, α, niter)
%timeit -r3 -n5 gd_numba(X, y, β, α, niter)
%timeit -r3 -n5 gd_cython(X, y, β, α, niter)
%timeit -r3 -n5 gd_cython2(X, y, β, α, niter)
```

```python
%%file test_arma.cpp

#include <iostream>
#include <armadillo>

using std::cout;
using std::endl;
        
int main() 
{
    using namespace arma;

    vec x = linspace<vec>(10,15,10);
    vec e = 10*randn<vec>(10);
    vec y;
    y = 3*pow(x,2)-7*x+2+e;
    vec xl = sqrt(x.t()*x);
    vec yl = sqrt(y.t()*y);
    vec d = x-y;
    vec dist = sqrt(d.t()*d);
    vec cov = (x-mean(x)).t()*(y-mean(y))/10;
    vec cor = cov/(stddev(x)*stddev(y));
    mat X;
    X = join_rows(ones(10),join_rows(x,pow(x,2)));
    vec b = inv(X.t()*X)*X.t()*y;
    
    cout << "\nVecotrs in Armadillo\n";
    cout <<"x:" << x << endl;
    cout <<"y:" << y << endl;
    cout <<"x's len:" << xl << endl;
    cout <<"y's len:" << yl << endl;
    cout <<"dist:" << dist << endl;
    cout <<"cor:" << cor << endl;
    cout <<"X:" << X << endl;
    cout <<"Coeff:"<< b <<endl;
}
```
